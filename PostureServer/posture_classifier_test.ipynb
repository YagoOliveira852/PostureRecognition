```python
import numpy as np
from json import load
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

```python
labels = [False, True]
true_labels = []

for i in range(10):
    true_labels = np.append(true_labels, labels)
```

```python
with open('logs.json', 'r') as logs_file:
    logs = load(logs_file)

predicted_labels = []

for log in logs[20:]: # Ãºltimas 20 fotos
    predicted_labels = np.append(predicted_labels, False if log["postura"] == "correta" else True)
```

```python
confusion = confusion_matrix(true_labels, predicted_labels)
ConfusionMatrixDisplay(confusion, display_labels=labels).plot()
```
